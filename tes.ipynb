```python
import cv2

# Inisialisasi objek video
video = cv2.VideoCapture(0)  # Angka 0 menandakan penggunaan kamera utama

# Tentukan format video output
fourcc = cv2.VideoWriter_fourcc(*"XVID")  # Format video XVID
output = cv2.VideoWriter(
    "output.avi", fourcc, 20.0, (640, 480)
)  # Nama file output, FPS, dan ukuran frame

# Loop untuk merekam video
while True:
    # Baca frame dari video
    ret, frame = video.read()

    if ret:
        # Tampilkan frame
        cv2.imshow("Video", frame)

        # Simpan frame ke video output
        output.write(frame)

        # Cek input tombol 'q' untuk keluar
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        break

# Tutup video dan jendela tampilan
video.release()
output.release()
cv2.destroyAllWindows()
```
